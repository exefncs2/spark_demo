```python
from pyspark.sql import SparkSession

# 創建 SparkSession
spark = SparkSession.builder \
    .appName("Jupyter PySpark Example") \
    .master("local[*]") \
    .getOrCreate()

# 創建 DataFrame
data = [("Alice", 34), ("Bob", 45), ("Catherine", 29)]
columns = ["Name", "Age"]

df = spark.createDataFrame(data, columns)

# 顯示 DataFrame 內容
df.show()

# 停止 SparkSession
spark.stop()
```
